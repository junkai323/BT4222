```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
full_clean_df = pd.read_excel("../data/full_clean_df.xlsx", index_col=0)
```

```python
labels_name = ['NotHate', 'Racist', 'Sexist', 'Homophobe', 'Religion', 'OtherHate']
```

```python
categories = labels_name
plt.figure(figsize=(10,6))

ax = sns.barplot(x=categories, y=full_clean_df.iloc[:, 1:7].sum().values)
plt.title("Tweets in each Category", fontsize=13)
plt.ylabel('Number of Tweets', fontsize=11)
plt.xlabel('Tweet Type ', fontsize=11)

#adding the text labels
rects = ax.patches
labels = full_clean_df.iloc[:,1:7].sum().values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom', fontsize=11)
    
plt.show()
```

```python
# Number of Categories for each tweet

rowSums = full_clean_df.iloc[:, 1:7].sum(axis=1)
multiLabel_counts = rowSums.value_counts()

plt.figure(figsize=(10,6))
ax = sns.barplot(x=multiLabel_counts.index, y=multiLabel_counts.values)

plt.title("Tweets having multiple labels", fontsize=13)
plt.ylabel('Number of Tweets', fontsize=11)
plt.xlabel('Number of Categories', fontsize=11)

#adding the text labels
rects = ax.patches
labels = multiLabel_counts.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
plt.show()
```
